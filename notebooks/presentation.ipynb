### Imports

```python
import os
import numpy as np
import pandas as pd
from PIL import Image 
import matplotlib.pyplot as plt
```

```python
def show_img(img_path, size=(11, 7)):
    plt.figure(figsize=size)
    plt.axis('off')
    plt.imshow(Image.open(img_path))
    plt.show()
```

## Let's look at PR (precision-recall) curve

```python
show_img(path_perfect_pr)
```

**Note:** The gray dotted line represents a “baseline” classifier — this classifier would simply predict that all instances belong to the positive class. The purple line represents an ideal classifier — one with perfect precision and recall at all thresholds. Nearly all real-world examples will fall somewhere between these two lines — not perfect, but providing better predictions than the “baseline”. A good classifier will maintain both a high precision and high recall across the graph, and will “hug” the upper right corner in the figure below.

## PR curve

**Note**: Important to notice - precission is greater than recall. In this type of tasks it's important to predict more false positives than false negatives e.g. it's better to see "one more car" and watch out for it than don't see fast car and crash with it.

Precision measures how accurate your predictions are. i.e. the percentage of your predictions are correct.

Recall measures how well you find all the positives.
